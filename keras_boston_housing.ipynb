# Normalize the values

It's important to use the `mean` and `standard deviation` from the *train* data to normalize the *test* data. Otherwise, if we used data from the training set the results would be biased.

```python
samples = np.random.randint(0, 11, size=1000).astype(np.float)
```

```python
samples.mean()
```

```python
samples.std()
```

```python

mean = samples.mean()
samples -= mean
std = samples.std()
samples /= std
```

# Build the model

```python
from keras import models
from keras import layers
```

```python
# input_shape = (train_data.shape[1], )
input_shape = (100, )
model = models.Sequential()
model.add(layers.Dense(64, activation='relu', input_shape=input_shape))
model.add(layers.Dense(1))
```

`mae` is mean absolute error

```python
model.compile(optimizer='rmsprop',
             loss='mse',
             metrics=['mae'])

```
